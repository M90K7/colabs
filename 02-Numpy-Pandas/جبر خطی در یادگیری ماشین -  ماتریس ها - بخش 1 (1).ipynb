 محاسبات برداری


1. جمع برداری

c = a + b   => c = (a1 + b1, a2 + b2, a3 + b3) 


```python
from numpy import array
# توصیف یک بردار
a = array([1, 2, 3])
print(a)
# بردار دوم
b = array([1, 2, 3])
print(b)
# جمع
c = a + b
print(c)
```

2. تفریق برداری

c = a − b   =>  c = (a1 − b1, a2 − b2, a3 − b3)

```python
from numpy import array

a = array([1, 2, 3])
print(a)

b = array([0.5, 0.5, 0.5])
print(b)

c = a - b
print(c)
```

3. ضرب برداری

c = a × b => c = (a1 × b1, a2 × b2, a3 × b3)

```python

from numpy import array

a = array([1, 2, 3])
print(a)

b = array([1, 2, 3])
print(b)

c = a * b
print(c)
```

4. تقسیم برداری

```python

from numpy import array

a = array([1, 2, 3])
print(a)

b = array([1, 2, 3])
print(b)

c = a / b
print(c)
```

5. ضرب داخلی

c = a · b =>  c = (a1 × b1 + a2 × b2 + a3 × b3)

```python

from numpy import array

a = array([1, 2, 3])
print(a)

b = array([1, 2, 3])
print(b)

c = a.dot(b)
print(c)
```

6. ضرب بردار اسکالر

### 2) نرم های برداری

 نرم برداری = محاسبه طول و یا اندازه بردار

 1.نرم بردار l1

L1(v) = ||v||1   //   ||v||1 = |a1| + |a2| + |a3|

```python
# نرم برداری l1
from numpy import array
from numpy.linalg import norm
# تعریف بردار
a = array([1, 2, 3])
print(a)
# محاسبه نرم
l1 = norm(a, 1) 
print(l1)
```

2. نرم بردار l2

L(v) = ||v||2 // ||v||2 = sqr ( a1^2 + a2^2 + a3^2)


```python
# نرم برداری l2
from numpy import array
from numpy.linalg import norm

a = array([1, 2, 3])
print(a)

l2 = norm(a)
print(l2)
```

3. نرم برداری ماکزیمم

Linf (v) = ||v||inf //  ||v||inf = max a1, a2, a3


```python
# نرم برداری ماکزیمم
from math import inf
from numpy import array
from numpy.linalg import norm

a = array([1, 2, 3])
print(a)

maxnorm = norm(a, inf)
print(maxnorm)
```

6. ماتریس متعامد

M. MT = I

```python

from numpy import array
from numpy.linalg import inv
# تعریف ماتریس
Q = array([
[1, 0],
[0, -1]])
print(Q)
# معکوس ماتریس
V = inv(Q)
print(Q.T)
print(V)
# identity equivalence
I = Q.dot(Q.T)
print(I)
```
